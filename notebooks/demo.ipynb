### Text to Speech

```python
from klaam import TextToSpeech
from IPython.display import Audio
import torch
import numpy as np
import os

root_path = "../klaam"
prepare_tts_model_path = os.path.join(root_path, "cfgs/FastSpeech2/config/Arabic/preprocess.yaml")
model_config_path = os.path.join(root_path, "cfgs/FastSpeech2/config/Arabic/model.yaml")
train_config_path = os.path.join(root_path, "cfgs/FastSpeech2/config/Arabic/train.yaml")
vocoder_config_path = os.path.join(root_path, "cfgs/FastSpeech2/model_config/hifigan/config.json")
speaker_pre_trained_path = os.path.join(root_path, "data/model_weights/hifigan/generator_universal.pth.tar")

# Load weights manually using weights_only=False
checkpoint = torch.load(speaker_pre_trained_path, weights_only=False, map_location="cpu")

# Initialize TTS
model = TextToSpeech(
    prepare_tts_model_path,
    model_config_path,
    train_config_path,
    vocoder_config_path,
    speaker_pre_trained_path=root_path  # leave path for klaam to find other files
)

# Synthesize text
model.synthesize("      -       ")
Audio("sample.wav")
```

```python
!pip install transformers soundfile

from transformers import pipeline

# Egyptian Arabic (arz)
pipe = pipeline("text-to-speech", model="facebook/mms-tts-arz")

text = "    "  # Egyptian dialect
speech = pipe(text)

with open("egyptian.wav", "wb") as f:
    f.write(speech["audio"])
```
